# Sort and Search

```python
# O(1) - Constant Time

def is_even(n):
  return n % 2 == 0

print(is_even(999))
```

```python
# O(n) - Linear Time Complexity

def count_up(n):
  for i in range(1, n+1):
    print(i)

print(count_up(10))
```

```python
# O(log n) - Logarithmic Time Complexity

def log_example(n):
  count = 1
  steps = 0
  while count < n:
    count *= 2
    steps += 1
    print(f"Step {steps} Count = {count}")

log_example(16)
```

```python
# O(n log n) - Quasilinear Time Comlexity

def split_half(n):
  if n <= 1:
    print(f"Reached {n}")
    return
  mid = n // 2
  print(f"Splitting {n} into {mid} and {n - mid}")
  split_half(mid)
  split_half(n - mid)

split_half(8)
```

```python
# O(n^2) - Quadratic Time Complexity

def print_pairs(n):
  for i in range(n):
    for j in range(n):
      print(f"({i}, {j})")

print_pairs(4)
```

```python
# O(n!) - Factorial Time Complexity

def print_permutations(n, used = [], count = None):
  if count is None:
    count = [0]

  if len(used) == n:
    print(used)
    count[0] += 1
    return

  for i in range(n):
    if i not in used:
      print_permutations(n, used + [i], count)

  if len(used) == 0:
    print({count[0]})

print_permutations(2)
```

```python
# LINEAR SEARCH - UNSORTED LIST

def linearSearch(theValues, target):
  n = len(theValues)
  for x in range(n):
    if theValues[x] == target:
      return x
  return -1


arr = [20, 10, 30, 50, 70]
target = 100

result = linearSearch(arr, target)

if result != -1:
  print(f"Linear Search: Target {target} found at index {result}")
else:
  print("Target not in dataset")
```

```python
# LINEAR SEARCH - SORTED LIST

def sortedLinearSearch(theValues, target):
  n = len(theValues)

  for x in range(n):
    if theValues[x] == target:
      return x
    elif theValues[x] > target:
      return -1
  return -1

arr = [10, 20, 30, 40, 50]
target = 100

result = sortedLinearSearch(arr, target)

if result != -1:
  print(f"Linear Search: Target {target} found at index {result}")
else:
  print("Target not in dataset")
```

```python
# BINARY SEARCH - SORTED LIST

def binarySearch(theValues, target):
  low = 0
  high = len(theValues) - 1

  while low <= high:
    mid = (high + low) // 2

    if theValues[mid] == target:
      return mid
    elif target < theValues[mid]:
      high = mid - 1
    else:
      low = mid + 1
  return -1

arr = [10, 20, 30, 40, 50, 60]
target = 30

result = binarySearch(arr, target)

if result != -1:
  print(f"Binary Search: Target {target} found at index {result}")
else:
  print("Target not in dataset")
```

```python
# BUBBLE SORT

def bubbleSort(arr):
  n = len(arr)

  for i in range(n):
    for j in range(0, n - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]

arr = [62, 34, 25, 12, 22, 11, 90]
bubbleSort(arr)
print("Bubble Sorted Array", arr)
```

```python
# SELECTION SORT

def selectionArray(arr):
  n = len(arr)

  for i in range(n):
    min_idx = 1
    for j in range(i+1, n):
      if arr[j] < arr[min_idx]:
        min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[j]

arr = [62, 34, 25, 12, 22, 11, 90]
selectionArray(arr)
print("Selection Sorted Array", arr)
```

```python
# INSERTION SORT

def insertionSort(arr):
  for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1

    while j >= 0 and key < arr[j]:
      arr[j + 1] = arr[j]
      j -= 1

    arr[j + 1] = key

arr = [62, 34, 25, 12, 22, 11, 90]
insertionSort(arr)
print("Insertion Sorted Array", arr)
```

```python
# QUICK SORT

def quickSort(arr):
  if len(arr) <= 1:
    return arr

  pivot = arr[len(arr) // 2]

  left = []
  middle = []
  right = []

  for x in arr:
    if x < pivot:
      left.append(x)
    elif x == pivot:
      middle.append(x)
    elif x > pivot:
      right.append(x)

  return quickSort(left) + middle + quickSort(right)

arr = [62, 34, 25, 12, 22, 11, 90]
quickSort(arr)
print("Quick Sorted Array", quickSort(arr))
```
